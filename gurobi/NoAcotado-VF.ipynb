```python
import gurobipy as gp
from gurobipy import GRB
```

```python
# Declara e inicializa el modelo
m = gp.Model('Modelo - No Acotado')
```

```python
# Crea las variables de decisión para el modelo Modelo1
x1 = m.addVar(name='x1')
x2 = m.addVar(name='x2')
# Integra las variables
m.update()
```

```python
m.addConstr(2*x1 + x2 >= 7, "r0")
m.addConstr(x1 + x2 >= 6, "r1")
m.addConstr(x1 + 3*x2 >= 9, "r2")
m.update()
```

```python
m.setObjective(3*x1 + 5*x2, GRB.MAXIMIZE)
m.update()
```

```python
m.display()
```

```python
m.optimize()
```

Debemos preguntar si es infactible o no acotado, en todo caso no existe solución óptima del modelo.

Pa ello utilizamos la variable del modelo **status**, que nos indica lo siguiente

*  ÓPTIMO (2), El modelo se resolvió de forma óptima (sujeto a tolerancias) y se encuentra disponible una solución óptima
*  INFACTIBLE (3)	Se demostró que el modelo no era factible
*  INF_OR_UNBD (4), se demostró que el modelo era inviable o no acotado. Para obtener una conclusión más definitiva, establezca el parámetro DualReductions en 0 y vuelva a optimizar
*  NO ACOTADO (5)	se demostró que el modelo no está acotado. Nota importante: un estado ilimitado indica la presencia de un rayo ilimitado que permite que el objetivo mejore sin límite. No dice nada acerca de si el modelo tiene una solución factible. Si necesita información sobre la viabilidad, debe establecer el objetivo en cero y volver a optimizar List item
