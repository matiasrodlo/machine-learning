```python
import gurobipy as gp
from gurobipy import GRB
```

```python
# Declara e inicializa el modelo
m = gp.Model('Modelo - Degenerado')
```

```python
# Crea las variables de decisión para el modelo Modelo1
x1 = m.addVar(name='x1')
x2 = m.addVar(name='x2')
s1 = m.addVar(name='s1')
s2 = m.addVar(name='s2')
s3 = m.addVar(name='s3')
# Integra las variables
m.update()
```

```python
m.addConstr(-3*x1 + 3*x2 + s1 == 9, "r0")
m.addConstr(2*x1 + x2    + s2 == 6, "r1")
m.addConstr(4*x1 + -3*x2 + s3 == 12, "r2")
m.update()
```

```python
m.setObjective(3*x1 - 2*x2, GRB.MAXIMIZE)
m.update()
```

```python
m.display()
```

```python
m.optimize()
```

```python
## Valor de la variables en el óptimo
for v in m.getVars():
    print(v.varName, ' = ', '{0:.2f}'.format(v.x))
```
