```python
import numpy as np

import gurobipy as gp
from gurobipy import GRB
```

```python
m = gp.Model('Sensibilidad-base')
```

```python
# Crea las variables de decisión para el modelo Modelo1
x1 = m.addVar(name='x1')
x2 = m.addVar(name='x2')
```

```python
m.addConstr(2*x1 + 4*x2 <= 20, "r0")
m.addConstr(8*x1 + 2*x2 <= 24, "r1")
m.addConstr(3*x1 + 2*x2 <= 12, "r2")
```

```python
m.setObjective(30*x1 + 40*x2, GRB.MAXIMIZE)
m.update()
```

```python
m.display()
```

```python
m.optimize()
```

```python
print('Sensitivity Analysis (SA)\nObjVal =', m.ObjVal)
m.printAttr(['X', 'Obj', 'RC', 'SAObjLow', 'SAObjUp'])
## Pi es el precio sombra (shadow price) que corresponde a la variable dual asociada
m.printAttr(['Sense', 'Slack', 'Pi', 'RHS', 'SARHSLow', 'SARHSUp'])
```
